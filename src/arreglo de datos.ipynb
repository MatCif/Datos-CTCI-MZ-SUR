A continuación se corrigen las ambiguedades en las instituciones de la mz sur

```python
#Se lee el dataset
path="https://raw.githubusercontent.com/ANID-GITHUB/Historico-de-Proyectos-Adjudicados/master/BDH_Proyectos.csv"
df=pd.read_csv(path, sep=";")

#Creación de diccionarios para el reemplazo de ambiguedades instituciones mz sur
#Universidades con casa central en las regiones de la araucanía, los ríos y los lagos
#Universidad Austral de Chile
institucion_dict_UACH={"UNIVERSIDAD AUSTRAL":"UNIVERSIDAD AUSTRAL DE CHILE","UNIV. AUSTRAL DE CHILE":"UNIVERSIDAD AUSTRAL DE CHILE",\
    "Universidad Austral de Chile":"UNIVERSIDAD AUSTRAL DE CHILE","JARDIN INFANTIL UNIVERSIDAD AUSTRAL DE CHILE":"UNIVERSIDAD AUSTRAL DE CHILE"} 
#Universidad Católica de Temuco
institucion_dict_UCT={"UNIV.CATOLICA DE TEMUCO":"UNIVERSIDAD CATOLICA DE TEMUCO"} 
#Universidad de la Frontera
institucion_dict_UFRO={"UNIV.DE LA FRONTERA":"UNIVERSIDAD DE LA FRONTERA","U DE LA FRONTERA":"UNIVERSIDAD DE LA FRONTERA"} 
#Universidad de Los Lagos
institucion_dict_ULAGOS={"UNIV.DE LOS LAGOS":"UNIVERSIDAD DE LOS LAGOS","U DE LOS LAGOS":"UNIVERSIDAD DE LOS LAGOS"} 

#Otras instituciones asentadas en la macrozona
#Centros
#Centro de estudios científicos
institucion_dict_centros={"CENTRO DE ESTUDIOS CIENTIFICOS (CECS)":"CENTRO DE ESTUDIOS CIENTIFICOS"\
    ,"CENTRO DE GENOMICA NUTRICIONAL AGROACUICOLA (CGNA)":"CENTRO DE GENOMICA NUTRICIONAL AGROACUICOLA"\
    ,"CENT. GENOM. NUT. AGROACUICOLA - CGNA":"CENTRO DE GENOMICA NUTRICIONAL AGROACUICOLA",\
    "CGNA":"CENTRO DE GENOMICA NUTRICIONAL AGROACUICOLA"} 

#Corrección de valores
df.replace({"INSTITUCION_PRINCIPAL": institucion_dict_UACH})
df.replace({"INSTITUCION_PRINCIPAL": institucion_dict_UCT})
df.replace({"INSTITUCION_PRINCIPAL": institucion_dict_UFRO})
df.replace({"INSTITUCION_PRINCIPAL": institucion_dict_ULAGOS})
df.replace({"INSTITUCION_PRINCIPAL": institucion_dict_centros})
```

Finalmente se exportan los datos corregidos a un registro nuevo
